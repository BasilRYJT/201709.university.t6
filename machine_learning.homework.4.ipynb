\end{aligned}
$$

# Question 4 

$$\begin{aligned}q(u,v)&=\frac{\text{count}(u,v)}{\text{count}(u)}\\e(u,v)&=\frac{\text{count}(u\rightarrow v)}{\text{count}(u)}\\p(u, x_1,x_2)&=\frac{\text{count}(u,x_1\rightarrow x_2)}{\text{count}(u,x_1)}\end{aligned}$$  

---

**Forward**  

$\begin{aligned}\alpha(u,1)&=\Pr(y_1=u)\\&=q(\text{start},u)\end{aligned}$  

$\begin{aligned}\alpha(u,2)&=\Pr(y_2=u)\\&=\sum_{v\in T}\alpha(v,1)\cdot q(v,u)\cdot e(v,x_1)\cdot p(v,x_1,y_2)\end{aligned}$  

$\begin{aligned}\therefore\alpha(u,k)&=\sum_{v\in T}\alpha(v,k-1)\cdot q(v,u)\cdot e(v,x_{k-1})\cdot p(v,x_{k-1},y_k)\end{aligned}$  

---

**Backward**  

$\beta(u,n)=q(u,\text{stop})\cdot e(u,x_n)\cdot p(u,x_n,y_n)$  

$\beta(u,n-1)=\sum_{v\in T}\beta(u,n)\cdot q(u,v)\cdot e(u,x_{n-1})\cdot p(u,x_{n-1},y_{n-1})$  

$\therefore\beta(u,k)=\sum_{v\in T}\beta(u,k+1)\cdot q(u,v)\cdot e(u,x_{k})\cdot p(u,x_{k},y_{k})$  

---

$O(nN^2)$ is the complexity of the algorithm. $N^2$ operations for operations on $N$ terms for every step, with the number of steps defined by sentence length, $n$.
